```python
import string
import pandas as pd
import nltk
#nltk.download('averaged_perceptron_tagger')
#nltk.download('punkt')
```

```python
filepath = 'data/harry_potter_1.txt'
```

# delete punctuations function

```python
def strip_punctuations(line):
    for character in string.punctuation:
        line = line.replace(character, '')
    return line
```

```python
word_count = {}
with open(filepath, 'r') as fi:
    # for each line in the file
    for line in fi:
        # remove the punctuation
        line = strip_punctuations(line)
        # then separate the words by spaces into a list
        words = line.split()

        # for each word in words
        for word in words:
            # convert the word into lowercase
            word = word.lower()
            # and check if it is NOT in or dictionary
            if word not in word_count:
                # if not, create a new entry for it
                word_count[word] = 0
            # Then, increment its count by 1
            word_count[word] += 1
```

```python
word_count[:10]
```

# Create pandas df from dict word_count

```python
word_count = pd.DataFrame([word_count]).transpose().reset_index().rename({'index':'word',0:'count'},axis=1)
```

# Create word tokens with nltk
https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html

```python
tokenized = list(word_count.keys())
tokenized = nltk.pos_tag(tokenized)
words_tags = dict(tokenized)
```

```python
words_tags
```

```python
tags = pd.DataFrame([words_tags]).transpose().reset_index().rename({'index':'word',0:'tag'},axis=1)
```

```python
df = pd.merge(word_count,tags,how='inner',on='word')
```
